# Data Visualization Using Python
Python offers several plotting libraries,namely Matplotlib,Seaborn and many other such data visualization packages with different features for creating informative,customized and appealing plots to present data in the most simple and effective way.

# Matplotlib and Seaborn 
Matplotlib and Seaborn are Python libraries that are used for data visualization.They have inbuilt modules for plotting different graphs.While Matplotlib is used to embed graphs into applications.Seaborn is primarily used for statistical graphs.

But when should we use either of the two ?Let's understand this with the help of a comparative analysis.Below are the comparison between Python's two well-known visualization packages Matplotlib and Seaborn.


## Matplotlib 
* It's used for basic graph plotting like line charts,bar graphs,etc
* It mainly works with datasets and arrays.
* Matplotlib acts productively with data arrays and frames.It regards the aces and figures as objects
* Matplotlib is more customizable and pairs well with Pandas and Numpy for Exploratory Data Analysis.


## Seaborn 
* It's mainly used for statistics visualization and can perform complex visualizations with fewer commands.
* It works with entire datasets.
* Seaborn is considerably more organized and functional than Matplotlib and treats the entire dataset as a solitary unit.
* Seaborn has more inbuilt themes and is mainly used for statistical analysis.

# Line Chart
A line chart is a graph that represents information as a series of data points connected by a straight line.In line charts,each data point or marker is plotted and connected with a line or curve.

Let's consider the apple yield(tons per hectare) in Kanto.Let's plot a line graph using this data and see how the yield of apples changes over time.We are going to both plot by using Matplotlib and Seaborn as follows:

## Using Matplotlib
We are using random data points to represent the yield of apples.

```python
yield_apples = [0.89,0.91,0.92,0.93,0.94,0.95] #ton per hectare
```

After we define the data points of apple yields,we are going to plot it's Line Chart using Matplotlib

```python
plt.plot(yield_apples)
```

To better understand the graph and it's purpose,we can add the x-axis values too.

```python
years = [2018, 2019, 2020,2021,2022]
yield_apples = [0.89, 0.91, 0.92, 0.94, 0.955] # ton per hectare
```

Since we have the 'years' variable as the x-axis,we need to type it before the yield_apples variable below

```python
plt.plot(years, yield_apples) # plt.plot(x,y)
```

Let's add labels to the axes so that we can show what each axis represents.

```python
plt.plot(years, yield_apples)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

salary =[0.8,0.9,0.5,0.5,0.8,0.5,0.3,0.6]
month=['Jan','Feb','March','April','Mei','Juni','July','August']

plt.plot(month, salary,color='black') # plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Salary(Millions$)')
plt.title("Salary in Month (Jan-Aug)",fontsize=15,fontweight='bold',color='#264653' )
plt.axvline(month[1],color='red')
plt.axvline(month[6],color='blue')
```

To plot multiple datasets on the same graph,just use the plt.plot function once for each dataset.Let's use this to compare the yields of apples vs oranges on the same graph

```python
years = [2017,2018, 2019, 2020,2021,2022]
apples =[0.89,0.91,0.93,0.97,0.986,0.943]
tangerines = [0.79,0.61,0.93,0.97,0.962,0.931]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.plot(years,apples)
plt.plot(years,tangerines)
plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")
```

We can add a legend which tells us what each line in our graph means.To understand what we are plotting,we can add a title to our graph.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.plot(years,apples)
plt.plot(years,tangerines)

plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','Tangerines'])
```

To show each data point on our graph,we can highlight them with markers using the marker argument.Many different marker shapes like a circle,cross,square,diamond,etc; are provided by Matplotlib.

```python
import seaborn as sns 
import matplotlib.pyplot as plt

plt.plot(years,apples,marker='o')
plt.plot(years,tangerines,marker='x')

plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','Tangerines'])
```

# Bar Graph 
When you have categorical data,you can represent it with a bar graph.A bar graph plots data with the help of bars,whick represent value on the y-axis and category on the x-axis.Bar graphs use bars with varying heights to show the data which belongs to a specific category.

## Using Matplotlib

```python
import matplotlib.pyplot as plt

years = [2017,2018, 2019, 2020,2021,2022]
apples =[0.89,0.91,0.93,0.97,0.986,0.943]
tangerines = [0.79,0.61,0.93,0.97,0.962,0.931]

plt.bar(years,tangerines) #plt.bar(x,y)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
```

```python
import matplotlib.pyplot as plt

years = [2017,2018, 2019, 2020,2021,2022]
apples =[0.89,0.91,0.93,0.97,0.986,0.943]
tangerines = [0.79,0.61,0.93,0.97,0.962,0.931]

plt.bar(years,apples) #plt.bar(x,y)
plt.bar(years,tangerines,bottom=apples) #plt.bar(x,y)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title("Crop Yields in Kanto")
```

## Using Seaborn
Let's use the tips dataset in Seaborn next.The dataset consists of : 
* Information about the sex (gender)
* Time of day
* Total bill
* Tips given by customer visiting the restaurant for a week

This Tips Dataset is a dataset provided by seaborn library,so you can directly load the dataset.

```python
import seaborn as sns

tips_df = sns.load_dataset("tips")
tips_df.head()
```

We can draw a bar chart to visualize how the average bill amount varies across diferent days of the week.We can do this by computing the day-wise averages and then using plt.bar.The Seaborn library also provides a barplot function that can automatically compute averages.

```python
import seaborn as sns

sns.barplot(x='day',y='total_bill',data=tips_df)
```

If you want to compare bar plots side-by-side,you can use the hue argument.The comparison will be done based on the third feature specified in this argument.


```python
import seaborn as sns
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

You can make the bars horizontal by switching the axes

```python
import seaborn as sns

tips_df = sns.load_dataset("tips")
sns.barplot(x='total_bill',y='day',hue='sex',data=tips_df)
```
